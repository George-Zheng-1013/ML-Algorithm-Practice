# k近邻算法

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("/content/drive/MyDrive/数据集/电影数据集.xlsx")
df.drop(columns="序号",inplace=True)
input_df=df.iloc[-1]
input_df=input_df.to_frame().T
df.drop(index=df.index[-1],inplace=True)
print(input_df)
print(df)
```

```python
def distance(df,input_df):
  搞笑dis=input_df['搞笑镜头']-df['搞笑镜头']
  搞笑dis=搞笑dis**2
  拥抱dis=input_df['拥抱镜头']-df['拥抱镜头']
  拥抱dis=拥抱dis**2
  打斗dis=input_df['打斗镜头']-df['打斗镜头']
  打斗dis=打斗dis**2
  距离=(搞笑dis+拥抱dis+打斗dis)**0.5
  return 距离.item()
```

```python
distances=[]
for index,row in df.iterrows():
  caculated_distance=distance(row,input_df)
  distances.append(caculated_distance)
df['距离']=distances
df=df.sort_values(by='距离')
print(df)
```

```python
K=5
nearest_neighbors=df.iloc[:K]
print(nearest_neighbors)
dict={}
for index,row in nearest_neighbors.iterrows():
  if row['电影类型'] not in dict:
    dict[row['电影类型']]=1
  else:
    dict[row['电影类型']]+=1
print(dict)
```

```python
电影类型=max(dict.items(),key=lambda x:x[1])[0]
print(f"属于{电影类型}")
```

# 使用scikit-learn实现

```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.neighbors import KNeighborsClassifier
```

```python
#构造数据集
x=[[0],[1],[2],[3]]
y=[0,0,1,1]
```

```python
#模型训练
estimator=KNeighborsClassifier(n_neighbors=2)
estimator.fit(x,y)
```

```python
print(estimator.predict([[1]]))
```
