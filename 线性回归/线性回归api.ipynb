```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data=[[80,86,84.2],
      [82,80,80.6],
      [85,78,80.1],
      [90,90,90],
      [86,82,83.2],
      [82,90,87.6],
      [78,80,79.4],
      [92,94,93.4]]
data=pd.DataFrame(data)
x=data.iloc[:,0:2]
y=data.iloc[:,2]
```

```python
estimator=LinearRegression()
estimator.fit(x,y)
#相关系数
print(estimator.coef_)
print(estimator.predict([[100,80]]))
```
