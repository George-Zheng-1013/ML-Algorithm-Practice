```python
import numpy as np
import pandas as pd
import sklearn as sk
```

```python
data=[[40920,8.326976,0.953952,3],[14488,7.153469,1.673904,2],[26052,1.441871,0.805124,1],[75136,13.147394,0.428964,1],[38344,1.669788,0.134296,1]]
dating_df=pd.DataFrame(data,columns=['milage','Liters','Consumtime','target'])
dating_df
```

```python
from sklearn.preprocessing import minmax_scale
def minmax_demo(data):
  print(data)
  data_scaled=minmax_scale(data[['milage','Liters','Consumtime']],feature_range=(2,3))
  print('归一化处理结果：\n',data_scaled)
  return None
```

```python
minmax_demo(dating_df)
```

```python
from sklearn.preprocessing import StandardScaler
def standardise_demo(data):
  print(data)
  scaler=StandardScaler()
  data_scaled=scaler.fit_transform(data[['milage','Liters','Consumtime']])
  print('标准化处理结果：\n',data_scaled)
  return None
standardise_demo(dating_df)
```
