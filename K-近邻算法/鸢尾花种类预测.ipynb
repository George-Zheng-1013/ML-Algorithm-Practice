# sklearn数据集获取

```python
import sklearn
from sklearn.datasets import load_iris,fetch_20newsgroups
```

```python
iris = load_iris()
print(iris)
```

```python
news=fetch_20newsgroups(subset='all')
#小数据集没有subset参数
#iris_test=load_iris(subset='test')
```

load和fetch返回的数据类型datasets.base.Bunch（字典格式）
*   data: 特征数据数组
*   target: 标签数组
*   DESCR: 数据描述
*   feature_names: 特证名（新闻数据、手写数字、回归数据集没有）
*   target_names: 标签名

```python
#数据集属性描述
print("特征值：\n",iris.data)
print("目标值：\n",iris['target'])
print("特征值名字：\n",iris.target_names)
print("描述：\n",iris.DESCR)
```

# seaborn数据可视化

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['species']=iris.target
iris_df
```

```python
def plot_iris(iris,col1,col2):
  sns.lmplot(x=col1,y=col2,hue='species',data=iris,fit_reg=False)
  plt.xlabel(col1)
  plt.ylabel(col2)
  plt.title('鸢尾花种类分布图')
  plt.show()
```

```python
plot_iris(iris_df,'petal width (cm)','sepal length (cm)')
```

# 数据集划分

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
print(x_train,'\n\n',x_test,'\n\n',y_train,'\n\n',y_test)
```
